## Exercise #1 <br>
#### Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.

```python
i = 0
num = 0
while num<1000:
    i += 1
    num = i**3
    print(num)
```

## Exercise #2 <br>
#### Get first prime numbers up to 100<br>
Using the Sieve of Eratosthenes method.<br>
List *nums* contains a bool values, which tell us if the number (represented by the index) is prime.<br>
numbers (indexes) *0* and *1* are included for convenience.<br>

```python
up_to = 100
nums = [True]*(up_to+1)
for i in range(2,len(nums)):
    if nums[i]:
        for j in range(i*2,len(nums),i):
            nums[j] = False
for i in range(2,len(nums)):
    if nums[i]:
        print(i)

```

## Exercise 3 <br>
#### Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors

```python
try:
    age_num = int(input("How old are you?"))
except:
    print("Age must be an integer.")
else:
    if age_num < 18:
        print("kids")
    elif 18 <= age_num <= 65:
        print("adults")
    else:
        print("senior")
```
